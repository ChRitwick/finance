Problem #1 

```python
# Present value of annuity = annual cash flow ( 1 - ( 1 / ( 1 + rate of interest )^time period ) ) / rate of interest
def cash_flow(pva,i,t):
  acf=(pva*i)/( 1 - ( 1 / ( 1 + i )**t ) )
  return acf
```

```python
pva=float(input("enter the Present value of annuity"))
i=float(input("enter the rate of interest in percentage"))
i=float(i/100)
t=float(input("enter the time period (number of years) :"))
answer=cash_flow(pva,i,t)
print("Annual cash flow is :",float(answer) )
```

Problem #4

```python
# PV = payment * (1- ((1 + growth_rate)/ (1 + discount_rate))** time_period) / (discount_rate - growth_rate)

def pv(fv,gr,i,t):
  pv=fv*(1-((1+gr)/(1+i))**t)/(i-gr)
  return pv
```

```python
fv=float(input("enter the future value"))
i=float(input("enter the rate of interest" ))
i=float(i/100)
gr=float(input("enter the growth rate"))
gr=float(gr/100)
t=float(input("enter the Time period"))
answer=pv(fv,gr,i,t)
print("Present value is :",float(answer) )
```

```python
# PV = payment * (1- ((1 + growth_rate)/ (1 + discount_rate))** time_period) / (discount_rate - growth_rate)

def pv(fv,i,t):
  pv=fv*(1-((1)/(1+i))**t)/(i)
  return pv

def b_payment(i,t,loan,pva):
  remaining=loan-pva
  return remaining * (1 + (i))**t
```

```python

loan=float(input("enter the Amount taken as loan" ))
m_payment=float(input("enter the Monthly payment"))
i = float(input("enter the rate of interest" ))
i=float(i/1200)
t=float(input("enter the time period (number of months) :"))

pva = pv(m_payment,i,t)
answer=b_payment(i,t,loan,pva)

print("baloon payment is :",answer)
```
