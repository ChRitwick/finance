```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_HA = pd.read_csv('/content/southwestern_HA.csv')
data_HA
```

```python
for col in data_HA.columns:
    data_HA[col] = data_HA[col].str.replace('$','')
    data_HA[col] = data_HA[col].str.replace('(','')
    data_HA[col] = data_HA[col].str.replace(')','')
    data_HA[col] = data_HA[col].str.replace(',','')
data_HA
```

```python
data_HA = data_HA.set_index('Consolidated Statement of Income (Loss) - USD ($) $ in Millions')
```

```python
data_HA.index.values
```

```python
plot = data_HA.astype(float).loc["NET INCOME LOSS"].plot.line()
plot = data_HA.astype(float).loc["Total operating revenues"].plot.line()
plot.legend(['Net Income', 'Revenue'])
plot.set_title('Quarterly Revenues and Net Income')


plt.ylabel('$ Billions')

plt.show()
```
